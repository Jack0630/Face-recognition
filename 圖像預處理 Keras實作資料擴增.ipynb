```python
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt
```

```python
#Dog Towel.png / lemon.jpg
import matplotlib.image as mpimg
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

l=[]
img = load_img(r"C:\Users\USER\Desktop\數據分析練習檔案\Dog Towel.png",target_size=(150,150))
x = img_to_array(img) 
l.append(x)
l=np.array(l)
```

```python
l.shape
```

```python
datagen = ImageDataGenerator(
        rotation_range=40, #旋轉角度0~180
        width_shift_range=0.2, #寬縮放比例
        height_shift_range=0.2, #高縮放比例
        shear_range=0.2, #順時鐘傾斜
        zoom_range=0.2, #縮放
        horizontal_flip=True, #隨機水平翻轉
        fill_mode='nearest') #填補空白

#reflect 反向補
#nearest 最近的補
#constant補指定數
#wrap 正向補
```

```python
i=0
for batch in datagen.flow(l ,batch_size=1):
    plt.figure(i)
    dog_plot=plt.imshow(array_to_img(batch[0]))
    i+=1
    if i% 3 == 0:
        break
```
